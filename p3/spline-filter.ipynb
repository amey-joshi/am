```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as ss

x = np.array([1, 2, 3, 2, 1, 0, -1, -2, -3, -2, -1, 0, 1])
x1 = np.convolve(x, x)
h = np.array([1, -2, 1])
h1 = np.convolve(h, h)

x_coeff = ss.cspline1d(x1)
h_coeff = ss.qspline1d(h1)

x_coeff_star_h_coeff = np.convolve(x_coeff, h_coeff)
sc = ss.bspline(np.linspace(-6, 6, 13), 3 + 2 + 1)

result = np.convolve(x_coeff_star_h_coeff, sc)
plt.plot(np.linspace(0, len(result), len(result)), result, '-')
plt.show()
# x_star_h = np.convolve(x, h)
```

```python

s1 = ss.bspline(np.linspace(-12, 12, 25), 3)
X = np.convolve(x1, s1)

s2 = ss.bspline(np.linspace(-2, 2, 5), 2)
H = np.convolve(h1, s2)
Y = np.convolve(X, H)
plt.plot(np.linspace(0, len(Y), len(Y)), Y, '-')
```
