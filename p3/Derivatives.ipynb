```python
import numpy as np
import matplotlib.pyplot as plt

x = [1, 2, 3, 1, 2, 3]
y = [1, 2, 1, 2, 1, 2]
h = [0, 0, 1, -2, 0, 0]

x_ip_y = np.multiply(x, y)
plt.plot([-2, -1, 0, 1, 2, 3], [0, 0, 0, 2, 0, 0], 'o')
plt.plot([-2, 3], [0, 0])
plt.title(r'$d(x, y)/dx[1]$')
plt.show()
print('Derivative of inner product with respect to an observation.')
```

```python
h_star_x = np.convolve(h, x)
h_adj_star_x = np.convolve(np.flip(h), x)
result = h_star_x + h_adj_star_x
plt.plot(np.linspace(-5, 5, 11), [0, 0, 0, -3, 0, 0, 0, 0, 0, 0, 0], 'o')
plt.plot([-5, 5], [0, 0])
plt.title(r'$d(h \star x, x)/dx[-2]$')
plt.show()
print('Derivative of inner product of a convolution and a signal')
print('with respect to an observation.')
```
