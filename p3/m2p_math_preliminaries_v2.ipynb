# 1. Numbers
This tutorial will sketch some of the advanced mathematical concepts that we will use in image processing.

A brief note on notation:

* $\mathbb{N}$: the set of all natural numbers. $\mathbb{N} = \{1, 2, 3, \ldots\} $ 
* $\mathbb{Z}$: the set of all integers. That is, $\mathbb{Z} = \{\ldots, -3, -2, -1, 0, 1, 2, 3, \ldots\}$ 
* Rational numbers are of the form $ x =\frac{a}{b}, a,b \in \mathbb{Z}, b\neq 0$. a is called $\color{blue}{\text{numerator}}$ and b is called  $\color{blue}{\text{denominator}}$. This set is countable. All $x$ digits can be listed e.g. 0.2345.
  - Note that: **0.333333...** is not rational number. 
* Irrational numbers cannot be written in the form $a,b \in \mathbb{Z}, b\neq 0$ e.g. $\sqrt2$ ,the ratio $\pi$ of a circle's circumference to its diameter, Euler's number $e$, and the golden ratio $\varphi$. They form an uncountable set.
* $\mathbb{R}$: the set of all real numbers. It includes rational and irrational number sets. 
* $\mathbb{C}$ the set of all complex numbers. 
   - $z = x + iy \in \mathbb{C}; x,y \in \mathbb{R}$, $i = \sqrt{-1}$, $i^2=-1$. $x$ is called the real part and $y$ the imaginary part of the complex number. It is common to write $x = \Re(z)$ and $y = \Im(z)$.
   - $\bar{z} = x - iy$ is called the $\color{blue}{\text{conjugate}}$  of $z$. Sometimes it is denoted by $z^\ast$.
   - The modulus of the complex number is an indication of its magnitude. It is defined as $|z| = z\bar{z} = x^2 + y^2$.




# 2. Linear algebra
Linear algebra is a study set of mathematical objects called [vectors](https://www.mathsisfun.com/algebra/vectors.html).  

<!--$\color{red}{\text{when explain vectors, please provide a visual explaination e.g. figures, 3Blue1Braun youtube channels has nice animation we can use},  }$-->

The simplest example of a vector is the one that defines a position in a plane. We commonly write it as: $\vec{r} = x\hat{i} + y\hat{j}$
 
Vectors can be represented by matrices. The set of all column matrices of the form $(x, y)^T$ also have the same properties that $\vec{r}$ had. In fact, there is a one-to-one correspondence:

\begin{equation}
\vec{r} = x\hat{i} + y\hat{j} \Leftrightarrow \vec{r} = (x, y)^T \tag{1}
\end{equation}

Yet another example of a vector is the set of all $2 \times 2$ matrices. Sum of two such matrices is a member of the set and so is a scalar multiple of a matrix. The scalar here is, once again, a real number.

The idea of vector in mathematics is far more general than that considered in elementary physics. Because many kinds of objects can be considered vectors, mathematicians do not bother to mark them using an arrow over their heads or using a bold font.

Now that we have given a few motivating examples of vectors, let us define a $\color{blue}{\text{vector space}}$: 
* The word *space* is a synonym for a set. 
* A vector space $V$ is a set with the following properties:
  1. If $v_1, v_2 \in V$ then $v_1 + v_2 \in V$. Some of two or more vectors is also a vector.
  2. There exists a unique element $0 \in V$ such that $v + 0 = v$ for all $v \in V$. $0$ is called the the **identity element**.
  3. For every $v \in V$, there exists an element $-v$ such that $v + (-v) = 0$. $-v$ is called the **inverse** of $v$.
  4. If $v_1, v_2, v_3 \in V$ then $v_1 + (v_2 + v_3) = (v_1 + v_2) + v_3$. That is, the addition operation is **associative**.
  5. For all $v_1, v_2 \in V$, $v_1 + v_2 = v_2 + v_1$. That is, the addition operation is **commutative**.
  6. For all $\alpha \in \mathbb{C}$ and $v \in V$ $\alpha v \in V$.
  7. For all $\alpha, \beta \in \mathbb{C}$ and $v \in V$, $(\alpha + \beta)v = \alpha v + \beta v$. Likewise, $\alpha(\beta v) = (\alpha\beta)v$.

These properties might appear a bit technical but they are necessary. 

We will not belabor much on these technicalities and will refer the reader to a classic such as [Halmos' Finite-dimensinal vector spaces](https://www.amazon.com/Finite-Dimensional-Vector-Spaces-Second-Mathematics/dp/0486814866).

We now move on to the concept of **dimensions**, motivating it through a few examples.

  - In order to specify a point in a line, we need one number telling how far it is away from zero and whether it is on the left or right of zero. Since a single number suffices, we say that a line is a one-dimensional object.
  - We need two numbers to identify a point on the plane. These two numbers are its coordiates. Since we need two numbers to identify a point uniquely on a plane we say that the plane is a two-dimensional object.
  - In order to identify a point on a surface of a sphere we need two numbers, its latitude and longitude. Therefore, the surface of a sphere is two dimensional.
  - An example of a three-dimensional object is a box. To tell where a point is in a box, we need to specify three numbers $(x, y, z)$ which are its coordinates with respect to an origin.

We can write a vector in such a plane as $\vec{r} = x\hat{i} + y\hat{j}$. If, instead, we want to represent a point as $(x, y)^T$, then we can write 
$$
\begin{pmatrix}x \\ y\end{pmatrix} = x\begin{pmatrix}1 \\ 0\end{pmatrix} + y \begin{pmatrix}0 \\ 1\end{pmatrix}.
$$
The special vectors $\{\hat{i}, \hat{j}\}$ or $(1, 0)^T, (0, 1)^T$ are called the basis vectors. An important property of the basis vectors is that they cannot be written in terms of each other. The precise mathematical term for this property is that they are **linearly independent**. Further, there are as many of them as the number of dimensions. In fact, the idea of dimensions is defined in terms of the number of linearly independent basis vectors.

The set of all vectors $(x, y)^T$ where $x, y \in \mathbb{R}$ is denoted by $\mathbb{R}^2$. If $x, y$ are complex numbers then we denote the set as $\mathbb{C}^2$. The idea can be generalized to any (finite) number of dimensions. Thus, $\mathbb{R}^n$ is the set of $n$-dimensional vectors of the form $(x_1, \ldots, x_n)^T$, where $x_1, \ldots, x_n \in \mathbb{R}$.

The way $(1, 0)^T, (0, 1)^T$ is a basis of $\mathbb{R}^2$ (and $\mathbb{C}^2$), the set of $n$ vectors
$$
\begin{pmatrix}1 \\ 0 \\ 0 \\ \vdots \\ 0\end{pmatrix}, \begin{pmatrix}0 \\ 1\\0 \\ \vdots \\ 0\end{pmatrix}, \ldots, \begin{pmatrix}0 \\ 0 \\ \vdots \\0\\ 1\end{pmatrix},
$$ is a basis of $\mathbb{R}^n$ (and $\mathbb{C}^n$). A vector space can have several bases although all of them have the same number of elements. The basis mentioned in this paragraph is called the *standard basis* and its members are usually denoted by $e_1, \ldots, e_n$.

Consider an $n$-dimensional vector space $\mathbb{C}^n$. Let $v_1 = (\alpha_1, \ldots, \alpha_n)$ and $v_2 = (\beta_1, \ldots, \beta_n)$ be two vectors in it. We define their *inner product* as 
$$
(v_1, v_2) = \sum_{i=1}^n \alpha_i\bar{\beta}_i.
$$
The inner product is a scalar. In this case it is a complex number. Recall that $\bar{\beta}_i$ is the complex conjugate of $\beta_i$.

We can readily infer the following properties if an inner product.

1. $(v_1, v_2) = \overline{(v_2, v_1)}$. The order of vectors *matters* in an inner product.
2. $(\alpha v_1, v_2) = \alpha(v_1, v_2)$.
3. $(v, v) \ge 0$ for all $v \in \mathbb{C}^n$.

The complex conjugate of a real number is itself. Therefore, the definition of inner product of vectors in $\mathbb{R}^n$ is
$$
(v_1, v_2) = \sum_{i=1}^n \alpha_i\beta_i.
$$
In this case the inner product is a real number and $(v_1, v_2) = (v_2, v_1)$. That is, the order of vectors in the inner product *does not matter* when the vectors are real.

The inner product $(v, v)$ is called the **norm** of a vectors. It is the same as the length of a vector in $\mathbb{R}^n$ or $\mathbb{C}^n$. A norm defined in terms of an inner product is sometime called a norm *induced* by an inner product. We shall come across this term when we come to Hilbert spaces.

A vector space in which a norm is defined is called a **normed vector space**.

Linear algebra restricts itself to finite dimensional vector spaces. Why? Because doing so guarantees that a sum like $\alpha_1 e_1 + \cdots + \alpha_n e_n$ is always defined. If we relax this requirement, we get into the domain of functional analysis.

# 3. Functional Analysis
Consider the set $P_n$ of polynomials of degree at most $n$. That is the set $P_n$ consists all polynomials of degrees $0, 1, \ldots, n$. It is clearly a vector space. For addition two polynomials of degree at most $n$ will give another polynomial of degree at most $n$ and multiplying a polynomial does not change its degree. We can also convince ourselves that the set $\{1, x, \ldots, x^n\} = \{x^k : 0 \le k \le n\}$ is a basis of $P_n$. Therefore, $P_n$ is a $n$-dimensional vector space. What if we consider the set $P$ of all polynomials? In that case the basis will be $\{x^k: k \in \mathbb{N}\}$. Thus, $P$ is an infinite dimensional vector space.

Quite a few things change when we move from finite number of dimensions to infinitely many dimensions. In the case of $P_n$, if $a_m, m = 0, 1, \ldots, n$ is *any* sequence of real numbers, the sum
$$
p(x) = \sum_{m=0}^n a_mx^m
$$ is always defined and is a member of $P_n$. However, when we relax the constraint of $n$ being finite, then
$$
p(x) = \sum_{m=0}^\infty a_m x^m
$$ 
may or may not converge. A linear combination of the basis vectors in the case of a finite dimensional vector space is always a well-defined vector. The same may not be true in the case of infinite dimensional spaces. Infinite dimensional vector spaces are the object of study in functional analysis.

We will touch upon only that part of functional analysis that we need to understand the Hibert space. It is a fascinating subject but we cannot cover it in one small note. A delightful survey of the historical development of functional analysis and an introduction to its basic ideas is written by <a href="Kreyszig1984">some of the early workers in the field</a>. Before we get to Hilbert space, we will review a few more concepts in mathematical analysis.



## Sequences
A sequence is function that maps natural numbers to real or complex numbers. That is, for every natural number $n$ there is a real or a complex number $x_n$. It common to write a sequence as $\{x_1, x_2, \ldots, \}$ or $\{x_n\}$. Let us consider a few examples.

- $\{x_n = 2^n\}$. It can also be written as $\{2, 2^2, 2^3, \ldots\}$.
- $\{x_n = (-1)^n\}$. We can write it as $\{-1, 1, -1, 1, \ldots\}$.
- $\{x_n = a\}$, where $a$ is a real or complex constant. Its terms are $\{a, a, \ldots\}$.
- $\{x_n = 3^{-n}\}$ whose terms are $\{1/3, 1/9, 1/27, 1/81, \ldots\}$.

The terms of sequences like $\{x_n = 2^n\}$ grow as $n$ increases. They are called **unbounded sequences**. The terms of the sequence $\{x_n = (-1)^n\}$, on the other hand, are always confined to $\{-1, 1\}$. It is an example of a **bounded sequence** because all of its terms are bounded by the numbers $-1$ and $1$. None of the terms ever cross the bounds. Similarly, the sequence $\{x_n = a\}$ is also bounded. The terms of the sequence $\{x_n = 3^{-n}\}$ diminish as $n$ increases. They get increasingly closer to $0$ as $n$ becomes larger and larger. But they never fall below zero. The sequence is also bounded because its terms are never greater than $1/3$ and lesser than $0$.

In the case of some sequences, as $n$ grows larger, the terms $x_n$ seem to get closer to a number $l$ called its limit. If $\{x_n = 3^{-n}\}$, we observe that the terms get ever closer to $0$. The number $0$ is called the limit of the sequence. The sequence $\{x_n = a\}$ also has a limit. It is $a$ itself. A sequence that has a limit is called a **convergent sequence**. 

On the other hand, the sequence $\{x_n = (-1)^n\}$ does not have a limit. This is because as $n$ increases, the terms keep oscillating between $-1$ and $+1$. Similarly, the sequence $\{x_n = 2^n\}$ also does not have a limit. As $n$ increases $x_n$ keeps doubling. Such sequences are called **divergent sequences**.

## Metric space
The word metric is related to measurement. In the context of metric spaces, it is related to distance. A metric tells how to compute distance between two elements of a set. If the set were a plane then one  way to measure distance between points $(x_1, y_1)$ and $(x_2, y_2)$ is the euclidean formula:

 $\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$.
 
 here is no one way to measure distances. An alternative is $|x_1 - x_2| + |y_1 - y_2|$. Whatever be the formula, the distance metric must satisfy the following reasonable properties:

1. Distance is never negative.
2. Distance between point $x$ and point $y$ is the same as that between point $y$ and $x$.
3. Distance between points $x$ and $z$ is no greater than the sum of distances between $x$ and $y$ and $y$ and $z$.

Let us express these ideas in mathematical terms. Let $X$ be a non-empty set. A metric $d$ is a function that maps a pair of points to the distance between them. That is, it maps $X \times X$ to $[0, \infty)$ and it has the following three properties:

1. $d(x, y) \ge 0$ for all $x, y \in X$. $d(x, y) = 0$ if an only if $x = y$.
2. $d(x, y) = d(y, x)$.
3. $d(x, z) \le d(x, y) + d(y, z)$.

We have merely expressed in mathematical terms the three properties written in plain language.



## Completeness
The idea of completeness is subtle. It is related to convergence of sequences of elements of a set. If $X$ is a metric space with a metric $d$ and if $\{x_1, x_2, \ldots \}$ is a sequence of elements in $X$ then it is said to converge to an element $l$ if for any $\epsilon > 0$ we can find $N \in \mathbb{N}$ such that for all $n \ge N$, $d(l, x_n) \le \epsilon$. The element $l$ is called the limit of the sequence and the sequence is said to converge to a limit. This definition guarantees us that if $l$ is a limit of a sequence then any neighborhood of $l$ has all but a finitely many points of the sequence. 

Consider the sequence $\{x_n = 2^{-n}\}$. Suppose we are given an $\epsilon > 0$. Then it is clear that for all $n > N = \log_2(1/\epsilon)$, $2^{-n} < \epsilon$. But $2^{-n} = |0 - 2^{-n}| = d(0, 2^{-n})$. Thus, for all $n > N = \log_2(1/\epsilon)$, $d(0, 2^{-n}) < \epsilon$. Therefore, the limit of this sequence is $0$.

A metric space $X$ is said to be a complete metric space if the limit of any convergent sequence of its elements is a member of $X$.

We will not get deeper into analysis than this. Should anyone be interested in knowing more, we recommed him an extremely readable book by <a hred="Kreyszig1978">E. Kreyszig</a>.



## Hilbert space
A Hilbert space is a complete, normed vector space.

We will state (not prove) a few examples of Hilbert spaces. All of them are of interest to signal processing.

1. The space $\mathbb{R}^n$ with the usual euclidean metric.
2. The space $\mathbb{C}^n$ with the metric $d(x, y) = ||x - y||$. The norm is the one induced by the inner product.
3. The space $L^2[a, b]$ of all square-integrable functions on the interval $[a, b]$. Note that the name of this space is $L^2[a, b]$. It is not to be interpreted as a quantity $L$ that is squared. It is a set of functions for which 
$$
\int_a^b |f|^2 dt
$$
is finite. Mathematically,
$$
L^2[a, b] = \left\{f:[a, b] \mapsto \mathbb{R}: \int_a^b |f|^2 dt \text{ exists.}\right\}.
$$
The inner product of real-valued functions $f$ and $g$ is defined as
$$
(f,g) = \int_a^b f(t)g(t)dt
$$
and the norm of a function $f$ is $||f|| = \sqrt{(f, f)}$. We encourage you to treat functions as if they were points in an abstract **function space**,  $(f, g)$ as the 'distance' between functions $f$ and $g$ and $||f||$ as the 'length' or the 'magnitude' of the function $f$. If the functions were complex valued, then the inner product is defined as
$$
(x, y) = \int_a^b x(y)\bar{y}(t)dt.
$$
4. The space $l^2$ of all square-summable functions. A sequence $\{a_n\}$ is square summable if $\sum_{n \ge 0}|a_n|^2$ exists. The inner product of sequences $a = \{a_n\}$ and $b = \{b_n\}$ is defined as 

\begin{equation}
(a, b) = \sum_{k=0}^\infty a_n\bar{b}_n \tag{1}
\end{equation}


 and the norm of a sequence is $||a|| = \sqrt{(a, a)}$. Note the great similarity between $L^2$ and $l^2$.
