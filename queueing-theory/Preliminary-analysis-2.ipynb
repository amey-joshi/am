```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('wait_prob.csv')
```

```python
dates_of_interest = ['2020-09-14', '2020-09-15', '2020-09-16', '2020-09-17', '2020-09-18']
data_of_interest = data.loc[data['Period'].isin(dates_of_interest) & data['nagents'] > 0]
```

```python
data_of_interest.shape
```

```python
data_of_interest.head()
```

```python
def show_data(date):
    fig, ax = plt.subplots(1, 3, figsize=(16, 8))
    data_of_interest[data_of_interest['Period'] == date][['Create-time', 'E']].plot(ax=ax[0])
    data_of_interest[data_of_interest['Period'] == date][['Create-time', 'nagents']].plot(ax=ax[1])
    data_of_interest[data_of_interest['Period'] == date][['Create-time', 'wait-prob']].plot(ax=ax[2])
    plt.suptitle(date)
    plt.show()
```

```python
for date in dates_of_interest:
    show_data(date)
```
